```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
from pymongo import MongoClient

client = MongoClient('localhost', 27017)
db = client.PGAStats

wells_fargo= db.wellsFargo
raw_data = pd.DataFrame(list(wells_fargo.find()))
```

```python
raw_data
```

```python
raw_data2017 = raw_data[raw_data.year!='2018'].copy()
```

```python
clean_data1 = raw_data2017[['r1','r2','total','name','year']].copy()
```

```python
clean_data1 = clean_data1[raw_data.r1!='--'].copy()
clean_data1 = clean_data1[raw_data.r2!='--'].copy()
clean_data1 = clean_data1[raw_data.total!='--'].copy()
```

```python
clean_data1[['r1','r2','total','year']]=clean_data1[['r1','r2','total','year']].astype(float)
```

```python
clean_data1.reset_index(drop=True)
clean_data1
```

```python
this_year = clean_data1.copy()
```

```python
prev_year = clean_data1.copy()
prev_year['year']=prev_year['year']+1
prev_year['avg_round']=prev_year['total']/4
```

```python
data1 = pd.merge(prev_year,this_year, on=['name','year'], how='inner')
```

```python
data1
```

```python
data1['r2_lt_avg'] = np.where(data1['r2_y']<=data1['avg_round'],1,0) 
data1['r1_lt_avg'] = np.where(data1['r1_y']<=data1['avg_round'],1,0) 
```

```python
data = data1[['r2_lt_avg','r1_lt_avg','avg_round']]
```

```python
data
```

```python
y = data['r2_lt_avg']
x1 = data[['r1_lt_avg','avg_round']]
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
# Get the regression summary
results_log.summary()
```

```python
def confusion_matrix(data,actual_values,model):
        
        # Confusion matrix 
        
        # Parameters
        # ----------
        # data: data frame or array
            # data is a data frame formatted in the same way as your input data (without the actual values)
            # e.g. const, var1, var2, etc. Order is very important!
        # actual_values: data frame or array
            # These are the actual values from the test_data
            # In the case of a logistic regression, it should be a single column with 0s and 1s
            
        # model: a LogitResults object
            # this is the variable where you have the fitted model 
            # e.g. results_log in this course
        # ----------
        
        #Predict the values using the Logit model
        pred_values = model.predict(data)
        # Specify the bins 
        bins=np.array([0,0.5,1])
        # Create a histogram, where if values are between 0 and 0.5 tell will be considered 0
        # if they are between 0.5 and 1, they will be considered 1
        cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
        # Calculate the accuracy
        accuracy = (cm[0,0]+cm[1,1])/cm.sum()
        # Return the confusion matrix and 
        return cm, accuracy
```

```python
confusion_matrix(x,y,results_log)
```
