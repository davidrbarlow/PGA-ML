```python
wf_data = wf_data[['year','name','score']]
```

```python

#wf_data['cut'] = np.where(wf_data['score'] in ['CUT','WD']).copy()
#wf_data['cut'] = wf_data['score'].where((wf_data['score']!='CUT') & (wf_data['score']!='WD'),1)
#wf_data['cut'] = wf_data['score'].where((wf_data['score']=='CUT') | (wf_data['score']=='WD'),0)
#np.where(data1['r2_y']<=data1['avg_round'],1,0)
wf_data['cut'] = 1
wf_data.loc[wf_data['score'] == 'CUT','cut'] = 0
wf_data.loc[wf_data['score'] == 'WD','cut'] = 0
wf_data.loc[wf_data['score'] == '--','cut'] = 0
```

```python
data1 = pd.merge(wf_data,expanded_data, on=['name','year'], how='inner')
```

```python
data1.drop(['_id'], axis=1)
```

```python
data1 = data1[['cut','driveAcc','gir','puttAvg','savePCT','yardPerDrive']].astype(float)
```

```python
estimators=['driveAcc','gir','puttAvg','savePCT','yardPerDrive']
from sklearn.model_selection import train_test_split

train, test = train_test_split(data1, test_size=0.2)
```

```python
X1_all = train[estimators]
y = train['cut']
```

```python
X_all = sm.add_constant(X1_all)
reg_logit = sm.Logit(y,X_all)
results_logit = reg_logit.fit()
results_logit.summary2()
```

```python
def confusion_matrix(data,actual_values,model):
        
        # Confusion matrix 
        
        # Parameters
        # ----------
        # data: data frame or array
            # data is a data frame formatted in the same way as your input data (without the actual values)
            # e.g. const, var1, var2, etc. Order is very important!
        # actual_values: data frame or array
            # These are the actual values from the test_data
            # In the case of a logistic regression, it should be a single column with 0s and 1s
            
        # model: a LogitResults object
            # this is the variable where you have the fitted model 
            # e.g. results_log in this course
        # ----------
        
        #Predict the values using the Logit model
        pred_values = model.predict(data)
        # Specify the bins 
        bins=np.array([0,0.5,1])
        # Create a histogram, where if values are between 0 and 0.5 tell will be considered 0
        # if they are between 0.5 and 1, they will be considered 1
        cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
        # Calculate the accuracy
        accuracy = (cm[0,0]+cm[1,1])/cm.sum()
        # Return the confusion matrix and 
        return cm, accuracy
```

```python
confusion_matrix(X_all,y,results_logit)
```

```python
y_test = test['cut']
# We already declared a list called 'estimators' that holds all relevant estimators for our model. 
X1_test = test[estimators]
X_test = sm.add_constant(X1_test)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# To make this as easy as possible to use, we declare a variable where we put
# all features where we want to check for multicollinearity
# since our categorical data is not yet preprocessed, we will only take the numerical ones
variables = data1[['driveAcc','gir','puttAvg','savePCT','yardPerDrive']]

```

```python
# we create a new data frame which will include all the VIFs
# note that each variable has its own variance inflation factor as this measure is variable specific (not model specific)
vif = pd.DataFrame()

# here we make use of the variance_inflation_factor, which will basically output the respective VIFs 
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
# Finally, I like to include names so it is easier to explore the result
vif["Features"] = variables.columns
```
